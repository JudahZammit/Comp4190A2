# Data Cleaning
1. We will cut out the first button press at the beginning of each data
2. We will cut out the last button press at the end of the Left data
3. We will make the normalize the time stamps such that the physical data starts at time 0

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
keys = ['Left','Right','Forward','Backward','Enter']
```

```python
trim = False
if(trim):
    cut_index = [125,135,225,150,125]
    left_cut_index = 100
else:
    cut_index = [0,0,0,0,0]
    left_cut_index = 0
```

```python
def cut(KEY,CUT):
        # Read in data
    gyro = pd.read_csv('./q2/myo_keyboard_data/'+KEY+'/gyro.csv',index_col = 'timestamp')
    acc = pd.read_csv('./q2/myo_keyboard_data/'+KEY+'/acc.csv',index_col = 'timestamp')
    orin = pd.read_csv('./q2/myo_keyboard_data/'+KEY+'/orin.csv',index_col = 'timestamp')
    eul = pd.read_csv('./q2/myo_keyboard_data/'+KEY+'/eul.csv',index_col = 'timestamp')
    emg = pd.read_csv('./q2/myo_keyboard_data/'+KEY+'/emg.csv',index_col = 'timestamp')
    
    cut_timestamp = gyro.index[CUT]
    if(trim):
        cut_gyro = gyro.loc[gyro.index > cut_timestamp]
        cut_acc = acc.loc[acc.index > cut_timestamp]
        cut_orin = orin.loc[orin.index > cut_timestamp]
        cut_eul = eul.loc[eul.index > cut_timestamp]
        cut_emg = emg.loc[emg.index > cut_timestamp]
    else:
        cut_gyro = gyro
        cut_acc = acc
        cut_orin = orin
        cut_eul = eul
        cut_emg = emg
        
    
    cut_gyro.to_csv('./q2/cleaned_data/denoised/'+KEY+'/gyro.csv')
    cut_acc.to_csv('./q2/cleaned_data/denoised/'+KEY+'/acc.csv')
    cut_orin.to_csv('./q2/cleaned_data/denoised/'+KEY+'/orin.csv')
    cut_eul.to_csv('./q2/cleaned_data/denoised/'+KEY+'/eul.csv')
    cut_emg.to_csv('./q2/cleaned_data/denoised/'+KEY+'/emg.csv') 
```

```python
for key,index in zip(keys,cut_index): cut(key,index)
```

```python
def left_cut(KEY = 'Left',CUT = 0):
    gyro = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/gyro.csv',index_col = 'timestamp')
    acc = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/acc.csv',index_col = 'timestamp')
    orin = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/orin.csv',index_col = 'timestamp')
    eul = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/eul.csv',index_col = 'timestamp')
    emg = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/emg.csv',index_col = 'timestamp')
    
    cut_timestamp = gyro.index[-CUT]
    if(trim):
        cut_gyro = gyro.loc[gyro.index < cut_timestamp]
        cut_acc = acc.loc[acc.index < cut_timestamp]
        cut_orin = orin.loc[orin.index < cut_timestamp]
        cut_eul = eul.loc[eul.index < cut_timestamp]
        cut_emg = emg.loc[emg.index < cut_timestamp]
    else:
        cut_gyro = gyro
        cut_acc = acc
        cut_orin = orin
        cut_eul = eul
        cut_emg = emg
    
    cut_gyro.to_csv('./q2/cleaned_data/denoised/'+KEY+'/gyro.csv')
    cut_acc.to_csv('./q2/cleaned_data/denoised/'+KEY+'/acc.csv')
    cut_orin.to_csv('./q2/cleaned_data/denoised/'+KEY+'/orin.csv')
    cut_eul.to_csv('./q2/cleaned_data/denoised/'+KEY+'/eul.csv')
    cut_emg.to_csv('./q2/cleaned_data/denoised/'+KEY+'/emg.csv') 
```

```python
left_cut(CUT = left_cut_index)
```

```python
def clean(KEY):
    gyro = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/gyro.csv',index_col = 'timestamp')
    acc = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/acc.csv',index_col = 'timestamp')
    orin = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/orin.csv',index_col = 'timestamp')
    eul = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/eul.csv',index_col = 'timestamp')
    emg = pd.read_csv('./q2/cleaned_data/denoised/'+KEY+'/emg.csv',index_col = 'timestamp')
    
    # Find the minimum timestamp
    MIN_TIMESTAMP = min(acc.index[0],emg.index[0])
    
    # Normalize timestamp
    gyro.index = gyro.index - MIN_TIMESTAMP
    acc.index = acc.index - MIN_TIMESTAMP
    orin.index = orin.index - MIN_TIMESTAMP
    eul.index = eul.index - MIN_TIMESTAMP
    emg.index = emg.index - MIN_TIMESTAMP
    
    gyro.to_csv('./q2/cleaned_data/normalized/'+KEY+'/gyro.csv')
    acc.to_csv('./q2/cleaned_data/normalized/'+KEY+'/acc.csv')
    orin.to_csv('./q2/cleaned_data/normalized/'+KEY+'/orin.csv')
    eul.to_csv('./q2/cleaned_data/normalized/'+KEY+'/eul.csv')
    emg.to_csv('./q2/cleaned_data/normalized/'+KEY+'/emg.csv')
    
```
